```python
import joblib
import pandas as pd
```

```python
model = joblib.load("model.pkl")
```

```python
print(f'''
Enter house detils:

1. House size in Square Foot
2. Land size in Perches
3. No of bed rooms
4. No of Bathrooms
5. Suburb no from following list
_____________
''')

df_suburbs = pd.read_csv('suburbs.csv')
for index, row in df_suburbs.iterrows():
    print(f'{index}. {row["suburbs"]}')
```

```python
# input data



house_size = 1000
land_size = 5
bed_rooms = 1
bath_rooms = 1
suburb_no = 22

```

```python
df_input = pd.DataFrame(columns = model.feature_names_in_)
df_input.loc[0] = None

if bed_rooms > 10:
    df_input['bedrooms'] = 10
    df_input['is_10_plus_bedrooms'] = 1
else:
    df_input['bedrooms'] = bed_rooms
    df_input['is_10_plus_bedrooms'] = 0

if bath_rooms > 10:
    df_input['bathrooms'] = 10
    df_input['is_10_plus_bathrooms'] = 1
else:
    df_input['bathrooms'] = bath_rooms
    df_input['is_10_plus_bathrooms'] = 0

df_input['house_size'] = house_size
df_input['land_size'] = land_size

column_name = 'suburb_' + df_suburbs.loc[suburb_no, 'suburbs']
if column_name in model.feature_names_in_:
    df_input.loc[0, column_name] = 1

df_input.fillna(0, inplace = True)

df_input
```

```python
pred_value = model.predict(df_input)[0]
print(f'Rs.{round(pred_value, 2)}')
```
